```python
#Iris Boxplot - Function
def boxplot(df):
    fig, axes = plt.subplots(2, 2, figsize = (12,10))

    sns.boxplot(data = df, y = 'SepalLength_Cm', x = 'Species', ax = axes[0,0]) #Boxplot - Sepal Length
    sns.boxplot(data = df, y = 'SepalWidth_Cm', x = 'Species', ax = axes[0,1]) #Boxplot - Sepal Width
    sns.boxplot(data = df, y = 'PetalLength_Cm', x = 'Species', ax = axes[1,0]) #Boxplot - Petal Length
    sns.boxplot(data = df, y = 'PetalWidth_Cm', x = 'Species', ax = axes[1,1]) #Boxplot - Petal Width

    plt.suptitle('Iris Dataset Boxplot') #Title
    plt.show()
    
boxplot(df)
```

```python
#Outlier Removal by IQR Method
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

# Calculate the interquartile range
IQR = Q3 - Q1

# Calculating lower and upper
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Outlier identification boundary
outliers = ((df < lower) | (df > upper)).any(axis=1)

# Outlier Removal
df_clean = df[~outliers]
```

```python
#Boxplot by cleaned data
boxplot(df_clean)
```

```python
#Comparing the rows / column of clean & original dataset
print("Original dataset shape:", df.shape)
print("Cleaned dataset shape:", df_clean.shape)
```

Model Creation

```python
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import svm #SVM Model 
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.metrics import confusion_matrix #Confusion Matrix
from sklearn import preprocessing #Label Encoder
from sklearn.preprocessing import StandardScaler #Scalar implementation
```

```python
#Feature and target separation
x = df_clean.drop('Species',axis = 1) #Feature
y = df_clean.Species #Target

#Species- from categorical to numerical. 0 = Iris-Setosa, 1 = Iris-Versicolor, 2 = Iris-virginica
le = preprocessing.LabelEncoder()
y = le.fit_transform(y)

print(y)
```

```python
#train and test split - test size = 30%, train size = 70%
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
#Scalear for SVM Model
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python

#Model Creation - SVM Model
iris_clas = svm.SVC(kernel = 'linear') #SVM Model
iris_clas.fit(x_train, y_train) #Fitting 
y_pred = iris_clas.predict(x_test) #Prediction 
```

```python
#Accuracy model calculation
accuracy = iris_clas.score(x_test,y_test)
accuracy_percent = round(accuracy * 100, 2) #Accuracy to two decimal points

print(f"Accuracy of the model: {accuracy_percent}%")
```

```python
plt.figure(figsize = (12,8))

#Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
names = ['Setosa', 'Versicolor', 'Virginica'] #Naming each species

#Confusion Matrix Heatmap
sns.heatmap(cm, annot = True, linewidths = 0.5, xticklabels=names, yticklabels=names, linecolor = "red", fmt = ' .0f')
plt.title('Confusion Matrix for Iris SVM Classifier')
plt.show()


#the model accurately predicted 17 instances,
#correctly identifying 10 Iris-Versicolor flowers and all 16 Iris-Virginica flowers. However,
#the model mistakenly classified one Iris-Versicolor flower as Iris-Virginica.
```

```python
from sklearn.metrics import classification_report

#Classification Report -
report = classification_report(y_test, y_pred, digits=3, output_dict=True)
df_report = pd.DataFrame(report).transpose()
print(df_report)

#perfect precision score for Iris-Setosa and Versicolor and .94 precision level for IrisVirginica
#recall score for Iris Setosa and Virginica is perfect with Versicolor at .90
#F1 Score, the harmonic mean of precision and recall are at perfect for Iris-Setosa, and at high of .95 for Iris-Versicolor and .96 for Iris-Virginic
#Model is predicting at a high precision accuracy.
```
